```python
import os
import pandas as pd

# Specify the directory path containing the folders
directory_path = "D:/downloads/GMT20240125-121850_RecordingnewChat/GMT20240125-121850_RecordingnewChat"

# Initialize lists to store data
all_timestamps = []
all_names = []
all_texts = []

# Loop through each folder in the directory
for folder_name in os.listdir(directory_path):
    # Construct the full folder path
    folder_path = os.path.join(directory_path, folder_name)
    
    # Check if the item in the directory is a folder
    if os.path.isdir(folder_path):
        # Loop through each file in the folder
        for file_name in os.listdir(folder_path):
            # Construct the full file path
            file_path = os.path.join(folder_path, file_name)
            
            # Read the text file line by line with UTF-8 encoding
            with open(file_path, 'r', encoding='utf-8') as file:
                for line in file:
                    # Split each line based on the assumed delimiter
                    parts = line.strip().split('\t')
                    # Extract data from each part assuming the format is timestamp, name, text
                    if len(parts) == 3:
                        all_timestamps.append(parts[0])
                        all_names.append(parts[1])
                        all_texts.append(parts[2])

# Create a DataFrame
df = pd.DataFrame({'timestamp': all_timestamps, 'name': all_names, 'text': all_texts})

# Display the DataFrame
print(df)
```
