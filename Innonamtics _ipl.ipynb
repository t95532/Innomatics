### toss winner win/lost the match by toss decision

```python
sns.countplot(data=matches, x='toss_decision/winner', hue='toss_decision')
```

```python
toss_winner = list(matches['toss_winner'])
toss_decision = list(matches['toss_decision'])
winner = list(matches['winner'])
```

```python
list2 = []
for i in range(0,len(winner)):
    if toss_winner[i]==winner[i]:
        if toss_decision[i]=='field':
            list2.append('won by chasing')
        else:
            list2.append('lost by chasing')
    else:
        if toss_decision[i] == 'bat':
            list2.append('lost by chasing')
        else:
            list2.append('won by chasing')
```

```python
len(list2)
```

```python
matches['chasing'] = list2
```

```python
matches.toss_decision.value_counts()
```
